# AI Impact Level ve Otomasyon Riski Dağılımı

🎯 Amaç:

- AI etkisinin dağılımını görmek (Low, Moderate, High)

- Otomasyon riski değerlerini histogramla incelemek

>**AI Impact Level Ne Demek?**
>
>AI Impact Level, bir mesleğin Yapay Zekâ teknolojilerinden ne düzeyde etkileneceğini ifade eder. Bu etki doğrudan o işin:
>
>- Otomatikleştirilebilirliği,
>
>- AI tarafından kolayca yapılabilirliği,
>
>- İnsan müdahalesine ne kadar ihtiyaç duyduğuyla ilgilidir.


>**Otomasyon Riski (%) Ne Demek?**
>
>- Bu, bir işin tamamen otomatik sistemler veya yapay zekâ tarafından yapılabilme olasılığıdır.
>- Yani yüzde olarak, "bu işin makinelerle değiştirilme ihtimali nedir" sorusuna cevap verir.

```python
sns.countplot(data=df, x="AI Impact Level", order=["Low", "Moderate", "High"])
plt.title("AI Impact Level Dağılımı")
plt.xlabel("AI Etkisi")
plt.ylabel("Meslek Sayısı")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="Automation Risk (%)", bins=30, kde=True)
plt.title("Otomasyon Riski (%) Dağılımı")
plt.xlabel("Otomasyon Riski (%)")
plt.ylabel("Meslek Sayısı")
plt.show()
```

# En Yüksek Otomasyon Riskine Sahip 10 Meslek

🎯 Amaç:

- Otomasyona en açık meslekleri öğrenmek

- Bu mesleklerin maaş, AI etkisi ve sektör gibi bilgilerini de göstermek

```python
top_10_risk = df.sort_values(by="Automation Risk (%)", ascending=False).head(10)

top_10_risk[["Job Title", "Industry", "Automation Risk (%)", "AI Impact Level", "Median Salary (USD)"]]
```

# Yüksek AI Etkisine Sahip Mesleklerde Maaş ve Uzaktan Çalışma Analizi
🎯 Amaç:

- "AI Impact Level" değeri "High" olan meslekler için

- Bu gruptaki mesleklerde:

- Ortalama maaş nedir?

- Uzaktan çalışabilme oranı nedir?

- En yüksek maaşa sahip ilk 10 meslek hangileri?

```python
high_ai = df[df["AI Impact Level"] == "High"]
high_ai.shape
```

```python
mean_salary = high_ai["Median Salary (USD)"].mean()
mean_remote = high_ai["Remote Work Ratio (%)"].mean()

print(f"Yüksek AI etkili mesleklerde ortalama maaş: ${mean_salary:,.2f}")
print(f"Uzaktan çalışabilme oranı ortalaması: %{mean_remote:.2f}")
```

```python
top10_salary_high_ai = high_ai.sort_values(by="Median Salary (USD)", ascending=False).head(10)
top10_salary_high_ai[["Job Title", "Industry", "Median Salary (USD)", "Remote Work Ratio (%)"]]
```

# Yüksek Maaşlı (High AI Impact) Meslekler

🎯 Amaç:


- En yüksek maaşlı meslekleri hem görsel hem metinsel olarak vurgulamak.

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top10_salary_high_ai,
    y="Job Title",
    x="Median Salary (USD)",
    palette="Blues_d"
)
plt.title("Yüksek AI Etkili Meslekler Arasında En Yüksek Maaşlı 10 Meslek")
plt.xlabel("Ortalama Yıllık Maaş (USD)")
plt.ylabel("Meslek")
plt.xlim(140000, 150000)
plt.show()
```

# Yüksek AI Etkisine Sahip Mesleklerin Sektörel Dağılımı

🎯 Amaç:

- Hangi sektörlerde AI etkisi yüksek seviyede olan meslekler daha fazla?

```python
sector_dist = high_ai["Industry"].value_counts().reset_index()
sector_dist.columns = ["Industry", "Count"]

plt.figure(figsize=(10, 6))
sns.barplot(data=sector_dist, x="Count", y="Industry", palette="coolwarm")
plt.title("Yüksek AI Etkili Mesleklerin Sektörel Dağılımı")
plt.xlabel("Meslek Sayısı")
plt.ylabel("Sektör")
plt.show()
```
